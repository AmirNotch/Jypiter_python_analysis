```python
# Задание 2

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataframe2015=pd.read_csv(r'C:\Users\ibrag\OneDrive\Рабочий стол/2015.csv')
dataframe2015['year']=2015
print('2015')
print(dataframe2015.columns)


dataframe2016=pd.read_csv(r'C:\Users\ibrag\OneDrive\Рабочий стол/2016.csv')
dataframe2016['year']=2016
print('2016')
print(dataframe2016.columns)


dataframe2017=pd.read_csv(r'C:\Users\ibrag\OneDrive\Рабочий стол/2017.csv')
dataframe2017['year']=2017
dataframe2017.rename(columns={'Economy..GDP.per.Capita.':'Economy (GDP per Capita)',\
                    'Trust..Government.Corruption.':'Trust (Government Corruption)',\
                    'Happiness.Score':'Happiness Score',\
                    'Health..Life.Expectancy.':'Health (Life Expectancy)','Dystopia.Residual':'Dystopia Residual','Happiness.Rank':'Happiness Rank'}, inplace=True)
print('2017')
print(dataframe2017.columns)

df_concat=pd.concat([dataframe2015,dataframe2016,dataframe2017])
df_concat.columns

top20_2015 = df_concat[df_concat.year==2015].iloc[:30]
top20_2016 = df_concat[df_concat.year==2016].iloc[:30]
top20_2017 = df_concat[df_concat.year==2017].iloc[:30]

plt.figure(figsize=(15,4))
plt.scatter('Country','Happiness Score', data=top20_2015,marker = "o", color='red')
plt.scatter('Country','Happiness Score', data=top20_2016,marker = "o", color='blue')
plt.scatter('Country','Happiness Score', data=top20_2017,marker = "o", color='green')
plt.xticks(rotation = 'vertical')
plt.legend(['2015','2016','2017'])
plt.title('Топ 25 странам с наибольшим абсолютным изменением индекса счастья 2015-2017', fontsize=14)

            
plt.show()
```
